```python
import pandas as pd
import requests
```

## step 1 : Make request to get the data

```python
response = requests.get('https://newsapi.org/v2/everything?q=tesla&from=2025-08-18&sortBy=publishedAt&apiKey=1f6519984b6743a48F627b11170a0770')
```

## Step 2: Convert response to JSON (dict)

```python
data = response.json()
data
```

## step 3

```python
# it Safely extracts the list of articles.
articles = data.get('articles', [])
```

## step 4: Convert to DataFrame

```python
# Convert to DataFrame
df = pd.json_normalize(articles)
```

## step 5: get selected cols

```python
df = df[['source.name', 'author', 'title', 'description', 'url', 'publishedAt', 'content']]
```
